```python
df = pd.read_csv("no_stem_reviews.csv")
```

```python
from sklearn.model_selection import train_test_split

X = df['review']
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=22,
    stratify=y
)
```

```python
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
import pandas as pd

train_df = pd.DataFrame({
    'review': X_train,
    'label': y_train})
```

```python
train_df['label'].value_counts()
```

```python
majority = train_df[train_df['label'] == 1]
minority = train_df[train_df['label'] == 0]
```

```python
print(len(majority))
print(len(minority))
```

```python
from sklearn.utils import resample

majority_downsampled = resample(
    majority,
    replace=False,
    n_samples=len(minority),
    random_state=22
)
```

```python
train_balanced = pd.concat([
    majority_downsampled,
    minority
])
```

```python
train_balanced['label'].value_counts()
```

```python
train_balanced = train_balanced.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)
```

```python
train_balanced.to_csv("undersampling_result.csv")
```
