```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

```

```python

df = pd.read_csv('no_stem_reviews.csv')

kolom_target = 'review' 

# Jika ada baris kosong
df[kolom_target] = df[kolom_target].fillna('')
```

```python

# Ubah kolom DataFrame kembali menjadi format list yang dipahami TfidfVectorizer
dokumen = df[kolom_target].tolist()
```

```python

dokumen[:5]
```

```python

# Inisialisasi model TF-IDF
vectorizer = TfidfVectorizer(lowercase=False)
```

```python


# Fit dan Transform
tfidf_matrix = vectorizer.fit_transform(dokumen)
```

```python

# sebagai kolom
kata_kunci = vectorizer.get_feature_names_out()
```

```python

print(f"Total kata unik (fitur) yang berhasil diekstrak: {len(kata_kunci)} kata")
print("=== Sampel 15 Kata Pertama ===")
print(kata_kunci[:15])

print("\n=== Sampel 15 Kata Terakhir ===")
print(kata_kunci[-15:])
```

```python


# Ubah hasil TF-IDF menjadi DataFrame
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=kata_kunci)
```

```python


print(f"Ukuran Dataset TF-IDF: {df_tfidf.shape[0]} baris dokumen x {df_tfidf.shape[1]} kolom kata\n")

print("=== Matriks Bobot TF-IDF (5 Baris Pertama) ===")
display(df_tfidf.head(5))

print("=== Matriks Bobot TF-IDF (5 Baris Tengah) ===")
display(df_tfidf.iloc[96:101])

print("=== Matriks Bobot TF-IDF (5 Baris Terakhir) ===")
display(df_tfidf.tail(5))
```

```python
df_tfidf['label'] = df['label']

df_tfidf.to_csv(
    'tfidf_nonstem.csv',
    index=False
)

print("Berhasil disimpan ke tfidf_nonstem.csv")
```
